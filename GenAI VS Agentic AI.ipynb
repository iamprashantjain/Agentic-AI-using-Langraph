# Whats GENAI?

Generative AI is a subset of AI focused on **learning the underlying distribution** of training data so it can generate **new, original samples** that resemble the original data.

**Timeline**
- Mainstream explosion: ~3 years (since late 2022 with ChatGPT)
- But foundational research (transformers, GANs, VAEs) goes back further


| Type | Models |
|------|--------|
| Text | ChatGPT, Claude, Gemini, Llama |
| Image | DALL·E, Midjourney, Stable Diffusion |
| Code | Codex, GitHub Copilot, CodeLlama |
| Audio/TTS | ElevenLabs, Bark |
| Video | Sora, Runway Gen-2, Pika |


**Traditional AI vs. GenAI – Key Difference**

- Traditional AI tries to find patterns in data, findout relationship b/w input & output and train model so that we can make predictions like spam classification model, image to cancer model, predicting used car price etc

- GenAI tries to find distribution of data so that it can generate a new sample from it

- basically, Traditional AI predicts an outcome based on past data whereas Generative AI creates new data that resembles the past data.


**Application Areas**
1. Creative and Buisness writing
2. Software development
3. Customer Support chatbots
4. Education
5. Designing


**GenAI VS Agentic AI**
- GenAI is constantly evolving and improving
- GenAI is about creating content, Agentic AI is about solving a goal
- GenAI is reactive whereas Agentic AI is proactive
- Agentic AI is GenAI with super powers
